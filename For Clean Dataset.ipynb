```python
import pandas as pd
import os

# The specific barcode you are looking for
target_barcode = "4001724036616"

file_paths = [
    r"C:\Users\xuyuh\Desktop\term3\Hackthon\en.openbeautyfacts.org.products.csv.gz",
    r"C:\Users\xuyuh\Desktop\term3\Hackthon\en.openfoodfacts.org.products.csv.gz",
    r"C:\Users\xuyuh\Desktop\term3\Hackthon\en.openpetfoodfacts.org.products.csv.gz",
    r"C:\Users\xuyuh\Desktop\term3\Hackthon\en.openproductsfacts.org.products.csv.gz"
]

def search_barcode(paths, target):
    found = False
    for path in paths:
        file_name = os.path.basename(path)
        print(f"Searching in {file_name}...")
        
        try:
            # Using chunksize to handle large files efficiently
            # We only load 'code' and 'product_name' columns to save memory
            chunks = pd.read_csv(
                path, 
                sep='\t', 
                encoding='utf-8', 
                usecols=['code', 'product_name'], # Add other columns if needed
                dtype={'code': str}, 
                compression='gzip',
                chunksize=100000,
                low_memory=False
            )
            
            for i, chunk in enumerate(chunks):
                # Search for the barcode in the 'code' column
                result = chunk[chunk['code'] == target]
                
                if not result.empty:
                    print(f"\n[MATCH FOUND!]")
                    print(f"File: {file_name}")
                    print(f"Chunk index: {i}")
                    print(result)
                    found = True
                    break # Stop searching this file
            
            if found: break # Stop searching other files if found
            
        except Exception as e:
            print(f"Error searching {file_name}: {e}")

    if not found:
        print(f"\nBarcode {target} was not found in any of the files.")

# Start the search
search_barcode(file_paths, target_barcode)
```

```python
import pandas as pd
import os

# 1. Path to your dataset
file_path = r"C:\Users\xuyuh\Desktop\term3\Hackthon\en.openfoodfacts.org.products.csv.gz"

# 2. White-list: The columns you want to KEEP
columns_to_keep = [
    "code", 
    "product_name", 
    "brands", 
    "categories", 
    "categories_tags", 
    "nutriscore_grade", 
    "environmental_score_grade",
    "nova_group"
]

def create_food_df_low_memory(path, keep_list):
    print("Starting low-memory chunked processing...")
    chunk_list = []
    
    try:
        # Step 1: Read in chunks of 50,000 rows
        # We load ALL columns, but only for a small number of rows at a time
        reader = pd.read_csv(
            path, 
            sep='\t', 
            encoding='utf-8', 
            dtype={'code': str}, 
            compression='gzip',
            low_memory=False,
            on_bad_lines='skip',
            chunksize=50000 
        )
        
        for i, chunk in enumerate(reader):
            # Step 2: Immediately drop unwanted columns from this chunk
            # This clears up memory for the next chunk
            cols_to_drop = [c for c in chunk.columns if c not in keep_list]
            chunk.drop(columns=cols_to_drop, inplace=True)
            
            # Step 3: Append only the slim version to our list
            chunk_list.append(chunk)
            
            if i % 10 == 0:
                print(f"-> Processed { (i+1)*50000 } rows...")

        # Step 4: Concatenate the slim chunks into one final DataFrame
        print("\nFinalizing the OPEN_FOOD_FACTS_FOOD dataframe...")
        df_final = pd.concat(chunk_list, ignore_index=True)
        
        # Ensure exact column order
        df_final = df_final[keep_list]
        return df_final

    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Execute the chunked process
OPEN_FOOD_FACTS_FOOD = create_food_df_low_memory(file_path, columns_to_keep)

# 3. Final Verification
if OPEN_FOOD_FACTS_FOOD is not None:
    print("\n" + "="*40)
    print("SUCCESS: Memory-safe processing complete!")
    print(f"Final Data Shape: {OPEN_FOOD_FACTS_FOOD.shape}")
    print("="*40)
    print(OPEN_FOOD_FACTS_FOOD.head())
```

```python
def analyze_barcode_quality(df):
    print("--- Barcode (code) Quality Analysis ---")
    
    # 1. Total row count
    total_rows = len(df)
    
    # 2. Number of unique barcodes
    # nunique() excludes NaN by default
    unique_count = df['code'].nunique()
    
    # 3. Calculate total duplicates
    duplicate_count = total_rows - unique_count
    
    # 4. Find the most frequent duplicate barcodes (Top 10)
    # value_counts() lists each code and how many times it appears
    duplicate_series = df['code'].value_counts()
    top_duplicates = duplicate_series[duplicate_series > 1].head(10)

    # Printing the Report
    print(f"Total Rows in Dataset:      {total_rows}")
    print(f"Unique Barcodes (Non-NaN):  {unique_count}")
    print(f"Duplicate Entries:          {duplicate_count}")
    print(f"Duplicate Percentage:       {(duplicate_count / total_rows) * 100:.2f}%")
    
    if not top_duplicates.empty:
        print("\n--- Top 10 Most Frequent Barcodes ---")
        print(top_duplicates)
    else:
        print("\nExcellent! No duplicate barcodes found.")

# Execute the analysis
analyze_barcode_quality(OPEN_FOOD_FACTS_FOOD)
```

```python
import pandas as pd

# 1. Define File Paths
receipt_clean_path = r"C:\Users\xuyuh\Desktop\term3\Hackthon\ESCP_DETAILED_RECEIPT_2026_CLEAN.parquet"
food_facts_path = r"C:\Users\xuyuh\Desktop\term3\Hackthon\OPEN_FOOD_FACTS_FINAL.parquet"

def finalize_database(r_path, f_path):
    print("Step 1: Loading Cleaned Files...")
    try:
        df_receipts = pd.read_parquet(r_path)
        df_food = pd.read_parquet(f_path)
        
        print(f"-> Receipts (Unique Products): {len(df_receipts)}")
        print(f"-> Food Facts Library: {len(df_food)}")

        print("\nStep 2: Performing Inner Join (Matching Barcodes)...")
        # Joining on upc_with_check_digit (Receipts) and code (Food Facts)
        merged_df = pd.merge(
            df_receipts,
            df_food,
            left_on='upc_with_check_digit',
            right_on='code',
            how='inner'
        )

        # Step 3: Cleanup redundant column
        if 'code' in merged_df.columns:
            merged_df.drop(columns=['code'], inplace=True)
            
        return merged_df

    except Exception as e:
        print(f"Error during final join: {e}")
        return None

# --- Execute the Join ---
ESCP_DETAILED_OPENFOODFACTS = finalize_database(receipt_clean_path, food_facts_path)

# --- Final Verification ---
if ESCP_DETAILED_OPENFOODFACTS is not None:
    print("\n" + "="*60)
    print("DATABASE: ESCP_DETAILED_OPENFOODFACTS")
    print(f"Total Matched Products: {len(ESCP_DETAILED_OPENFOODFACTS)}")
    print("="*60)

    # 1. Check Random 5 Rows
    print("\n--- [Check 1] Random 5 Rows Sample ---")
    print(ESCP_DETAILED_OPENFOODFACTS.sample(5))

    # 2. Check Specific Barcode: 4001724036616
    print("\n--- [Check 2] Searching for Pizza Barcode 4001724036616 ---")
    target_barcode = "4001724036616"
    pizza_check = ESCP_DETAILED_OPENFOODFACTS[ESCP_DETAILED_OPENFOODFACTS['upc_with_check_digit'] == target_barcode]
    
    if not pizza_check.empty:
        print("MATCH FOUND IN FINAL DATABASE:")
        print(pizza_check)
    else:
        print(f"Barcode {target_barcode} was NOT found in the joined results.")
        print("Note: This means the barcode exists in receipts, but is missing from your Open Food Facts file.")
```
